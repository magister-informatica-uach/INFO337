### Definición Formal de Probabilidad
Una Probabilidad $P$ es una función que cumple:
\begin{equation}
\begin{array}{cc} P:& {\cal P}(\Omega) \to [0,1]\\
&A \to P(A)
\end{array}
\end{equation}
que cumple lo siguiente:

$\begin{equation}
\begin{array}{ll} 1. & \forall A \subset \Omega, P(A) \geq 0\\
2. &  P(\Omega) = 1\\
3. & Sean \,\, A_1,...,A_n \subset \Omega: A_i \cap A_j =\phi, \forall i,j=1...n => P(A_1\cup...\cup A_n) = P(A_1)+...+P(A_n)\\
\end{array}
\end{equation}$

### Propiedades
$\begin{equation}
\begin{array}{ll} 
1. & P(A) = 1 - P(A^C)\\
2. &  P(\phi) = 0\\
3. & Si A \subset B => P(A) \leq P(B)\\
4. & P(A) \leq 1\\
5. & \forall A,B \in \Omega\,\, P(A \cup B) = P(A) + P(B) - P(A \cap B)
\end{array}
\end{equation}$

### ¿Cómo asignamos valores de probabilidad?

- $\textbf{Enfoque frecuentista relativo:}$ Sea $n$ el número de experimentos, y $n(A)$ el número de veces que el evento $A$ ocurre en la realización de esos experimentos, entonces $P(A) = \frac{n(A)}{n}$

- $\textbf{Enfoque frecuentista clásico:}$ Sea $n(A)$ el número elementos en el evento $A$, y $n(\Omega)$ el número de elementos en el espacio muestral $\Omega$, entonce $P(A) = \frac{n(A)}{n(\Omega)}$

### Probabilidades condicionales
Sean $A, B \subset \Omega$, si $P(B)>0$, se define:

$\begin{equation}
\begin{array}{lll} 
P(A\mid B) & = &\frac{P(A \cap B)}{P(B)}
\end{array}\end{equation}$

Si además se cumple que $P(A)>0$, se tiene:

$\begin{equation}
\begin{array}{lll}
P(A\cap B)& = &P(A\mid B)P(B) = P(B\mid A)P(A)
\end{array}\end{equation}$

Esto se puede extender a mas eventos:

$\begin{equation}
\begin{array}{lll}
P(A\cap B \cap C)& = &P(A)P(B\mid A) P(C\mid A \cap B) 
\end{array}\end{equation}$
siempre que $P(A), P(A\cap B)> 0$

### Independencia
¿Cómo determinar que dos eventos son independientes?

Sean $A, B \subset \Omega$ con $P(A),P(B) >0$, si cualquiera de las siguientes condiciones se cumple:

$\begin{equation}
\begin{array}{llll}
1. &P(A\cap B)& = &P(A)P(B)\\ 
2. &P(A\mid B) & = & P(A)\\
3. &P(B\mid A) & = & P(B)\\
\end{array}\end{equation}$

se dice que $A$ es independiente de $B$.

### Independencia Condicional
¿Cómo determinar que dos eventos $A$ y $B$ son independientes dados un tercer evento $C$?

Sean $A, B, C \subset \Omega$ con $P(A\cap C),P(B\cap C), P(C)>0$, si cualquiera de las siguientes condiciones se cumple:

$\begin{equation}
\begin{array}{llll}
1. &P(A\cap B \mid C)& = &P(A\mid C)P(B\mid C)\\ 
2. &P(A\mid B \cap C) & = & P(A \mid C)\\
3. &P(B\mid A \cap C) & = & P(B \mid C)\\
\end{array}\end{equation}$

se dice que $A$ es independiente de $B$ condicionalmente a $C$.

### Ley de las Probabilidades totales
Consideremos ahora una partición del espacio muestral $\{B_i\}_{i=1,..n}$, es decir:

$\begin{equation}
\begin{array}{lll}
\bigcup_{i=1}^n \limits  B_i = \Omega, & & B_i\cap B_j = \phi \qquad \forall i,j=1,..n, i\neq j
\end{array}\end{equation}$

entonces:

$\begin{equation}
\begin{array}{lll}
P(A) & = & \sum_{i=1}^n \limits P(A\mid B_i) P(B_i)
\end{array}\end{equation}$

Ejemplo: Sea $A$ = \{Pedro acepta trabajo en Chile\}
y $B_i$ = \{Oferta de trabajo para Pedro en region $i$ \}, $i=1,...,15$ 






### Regla de Bayes
A partir de la defición de probabilidades condicionales se puede deducir la regla de Bayes:

Sean $A, B \subset \Omega$ tales que $P(A),P(B)>0$, entonces 

$\begin{equation}
\begin{array}{lll}
P(B\mid A)& = & \frac{P(A\mid B) P(B)}{P(A)}
\end{array}\end{equation}$

$P(B)$ se denomina probabilidad a priori, $P(B \mid A)$ probabilidad a posteriori

### Teorema de Bayes
Consideremos una partición del espacio muestral $\{B_i\}_{i=1,..n}$, entonces:

$\begin{equation}
\begin{array}{lll}
P(B_i \mid A) & = & \frac{P(A\mid B_i) P(B_i)}{P(A)}
& = & \frac{P(A\mid B_i) P(B_i)}{\sum_{i=1}^n \limits P(A\mid B_i) P(B_i)}
\end{array}\end{equation}$

### Variables Aleatorias
Una variable aleatoria es una función que asigna un número a cada elemento en el espacio muestral de un experimento aleatorio. 
Se denomina Variable Aleatoria Discreta si los valores son enteros:

$\begin{equation}
\begin{array}{cc} X:& \Omega \to \mathbb{N}\\
&\omega \to X(\omega)
\end{array}
\end{equation}$

y Variable Aleatoria Continua si los valores son reales:

$\begin{equation}
\begin{array}{cc} X:& \Omega \to \mathbb{R}\\
&\omega \to X(\omega)
\end{array}
\end{equation}$


Ejemplos: 

$\begin{equation}
\begin{array}{ll} 
X:& \Omega \to \{0,1\}\\
&\omega \to X(\omega)= 
\left\{\begin{array}{ll} 1 & \omega=cara\\
                          0 & \omega=sello \\
\end{array} \right .\\
&\\
 X:& \Omega \to [0,\infty) \\
&\omega \to X(\omega)= t \\
\end{array}
\end{equation}$

donde $t$ es el tiempo de falla de una máquina y $\omega$ es el evento de falla con todas sus características no necesariamente observables.

### Función de probabilidad puntual
En el caso de variables aleatorias discretas, la probabilidades se definen mediante una función de probabilidad puntual.

Ejemplo: 

$\begin{equation}
\begin{array}{ll} 
X:& \Omega \to \mathbb{N}\\
&\omega \to X(\omega)= n \\
\end{array}
\end{equation}$

dónde n representa el número de clientes que llegan al banco el lunes entre 9 y 10 am 
$$P(X=10) = f(10) = ?$$

```r
##caso en que el nro de clientes se modela como Poisson de media 10
options(repr.plot.width=6, repr.plot.height=6)
vec <- seq(1,30,by=1)
pvec <- dpois(vec,10)
plot(vec,pvec,type="h",col="blue", xlab="n",ylab="P(X=n)")
points(vec,pvec,col="blue")
```

### Función de densidad de probabilidad
En el caso de variables aleatorias continuas, las probabilidades se definen mediante una función de densidad de probabilidad:
$\begin{equation}
\begin{array}{ll} 
f_X:& \mathbb{R} \to \mathbb{R_+}\\
&x \to f_X(x) \\
\end{array}
\end{equation}$

Ejemplo: 

$\begin{equation}
\begin{array}{ll} 
X:& \Omega \to \mathbb{R}\\
&\omega \to X(\omega)= x \\
f_X(x)& = \frac{1}{2\sqrt{2\pi}} exp(\frac{-(x-60)^2}{8})
\end{array}
\end{equation}$

dónde x representa el peso de un individuo.
$$P(58 <= X <= 62) = \int_{58}^{62} f(x) dx = F_X(62) - F_X(58) = \int_0^{62} f(x)dx - \int_0^{58} f(x)dx ?$$


```r
options(repr.plot.width=6, repr.plot.height=6)
vec <- seq(50,70,by=0.1)
par(cex=0.8)
plot(vec,dnorm(vec,60,2),type="l", xlab = "Peso", ylab = "Densidad", main = "Función de Densidad de Probabilidad")
densidad = dnorm(vec,60,2)
vec2 <- seq(58,62,by=0.1)
abline(h=0)
polygon(x = c(58,vec2,62), y =c(0,dnorm(vec2,60,2),0), col = "grey")
prob = pnorm(62,60,2)-pnorm(58,60,2)
print(prob)

     
```

### Función de distribución de probabilidad acumulada
Es la función $F_X$ que devuelve la probabilidad de que una v.a. sea menor o igual que un valor.

En el caso de las v.a. discretas se calcula como:

$\begin{equation}
\begin{array}{ll} 
P(X \leq x) = F_X(x) & = \sum_{t=1}^x \limits P(X = t)
\end{array}
\end{equation}$


En el caso de las v.a. continuas se calcula como:

$\begin{equation}
\begin{array}{llll} 
P(X \leq x)  = &F_X(x)& = &\int_{-\infty}^x f(x)dx\\
&f_X(x)& = & \frac{d}{dx} F_X(x)
\end{array}
\end{equation}$

Y una manera alternativa de obtener la probabilidad:

$\begin{equation}
\begin{array}{l}
P(a \leq X \leq b) = F_X(b) - F_X(a)\\
\end{array}
\end{equation}$

```r
par(mfrow=c(1,2))
options(repr.plot.width=10, repr.plot.height=5)
##distribución Poisson acumulada
vec <- seq(1,30,by=1)
pvec <- ppois(vec,10)
plot(vec,pvec,type="h",col="blue", xlab="n",ylab="F(n)",main="Probabilidad Acumulada: Caso Discreto")
points(vec,pvec,col="blue")
##distribución normal acumulada
vec <- seq(50,70,by=0.1)
plot(vec,pnorm(vec,60,2),type="l", xlab = "Peso", ylab = "Probabilidad", main = "Probabilidad Acumulada: Caso Continuo")
lines(vec,pnorm(vec,60,8),col="blue")
lines(vec,pnorm(vec,60,4),col="red")
abline(h=0)
legend(x = "topleft", legend = c("s= 2", "s= 4","s = 8"), lty=1 ,col = c("black", "red", "blue"))
```

### Variables Aleatorias Multidimensionales
Se trata de funciones que asignan un vector multidimensional al resultado de un experimento aleatorio. 
Se denomina Vector Aleatorio Discreto si los valores son enteros multidimensionales:

$\begin{equation}
\begin{array}{cl} {\bf X}:& \Omega \to \mathbb N^k\\
&\omega \to {\bf X}(\omega)=(X_1(\omega),\cdots,X_k(\omega)) = (n_1,\cdots,n_k)
\end{array}
\end{equation}$

y Vector Aleatorio Continuo si los valores son reales multidimensionales:

$\begin{equation}
\begin{array}{cl} {\bf X}:& \Omega \to \mathbb R^k\\
&\omega \to {\bf X}(\omega)=(X_1(\omega),\cdots,X_k(\omega)) = (x_1,\cdots,x_k)
\end{array}
\end{equation}$

#### Ejemplo:
Sea 

$\begin{equation}
\begin{array}{cl}(X,Y):&\Omega \to \mathbb R_{+}^2\\
&\omega \to (X(\omega),Y(\omega))=(x,y)
\end{array}
\end{equation}$

tal que su densidad de probabilidad es:
$$f(x,y) = ab \exp^{-(ax+by)} \qquad a,b>0$$
y entonces la distribución de probabilidad acumulada queda:
$$F(x,y)= \int_0^x \int_0^y f(x,y) dydx
= (1- \exp^{-ax})(1- \exp^{-by})$$
Notar que 
$$\lim_{x,y \to \infty} F(x,y) = 1$$


```r
options(repr.plot.width=10, repr.plot.height=5)
par(mfrow=c(1,2))
x <- seq(0,1,by=0.05)
y <- seq(0,1,by=0.05)
a=5
b=1
dens <- function(x,y) a*b*(exp (-a*x -b*y))
z = outer(x,y,dens)
persp(x,y,z,ticktype = "detailed",zlab="f(x,y)",main="Densidad de Prob. Bivariada: Caso a=5 y b=2")

dist <- function(x,y) 1-exp (-a*x -b*y) 
z = outer(x,y,dist)
persp(x,y,z,ticktype = "detailed",zlab="F(x,y)",main="Distribución  Bivariada: Caso a=5 y b=2")
```

### Esperanza
La esperanza o  valor esperado de una v.a. $X$ se denota $E[X]$ y se calcula como:
$\begin{equation}
\begin{array}{ll} 
E[X] = 
\left\{\begin{array}{ll} \sum_i x_i P(X=x_i) & si\,X\, discreta\\
                          \int x f_X(x)dx & si\,X\, continua\\
\end{array} \right .\\
\end{array}
\end{equation}$


Consideremos $g$ una función a valores reales, entonces:

$\begin{equation}
\begin{array}{lll}
E[g(X)] & = & 
\left\{\begin{array}{ll} \sum_i g(x_i) P(X=x_i) & si\,X\, discreta\\
                          \int g(x) f_X(x)dx & si\,X\, continua\\
\end{array}\right .\\
\end{array}
\end{equation}$

Para el caso especial de $g(x) = x^n$ se define el n-ésimo momento de X como:

$\begin{equation}
\begin{array}{lll} 
E[X^n] & = & 
\left\{\begin{array}{ll} \sum_i x_i^n P(X=x_i) & si\,X\, discreta\\
                          \int x^n f_X(x)dx & si\,X\, continua\\
\end{array}\right .\\
\end{array}
\end{equation}$

La esperanza es el primer momento y se denota $\mu$.

#### Propiedades
Sean $a,b \in \cal{R}$ entonces:

$\begin{equation}
\begin{array}{lll} 
E[aX+b] & = & aE[X] + b \\
E[X + Y] & = & E[X] + E[Y]\\
\end{array}
\end{equation}$



### Varianza y covarianza
La varianza mide la variación de la v.a. entorno a la esperanza o media $\mu$, y se define como

$\begin{equation}
\begin{array}{ll} 
Var(X) = E[(X-\mu)^2] = E[X^2] - \mu^2
\end{array}
\end{equation}$

Se cumple que:

$\begin{equation}
\begin{array}{ll} 
Var(aX+b) = a^2 Var(X)
\end{array}
\end{equation}$

Se define además la desviación estándar $\sigma = \sqrt{Var(X)}$ 


La covarianza mide la relación (lineal) que hay entre dos v.a. $X$ e $Y$. Si denotamos $\mu_X = E[X]$ y $ \mu_Y= E[Y]$ entonces:

$\begin{equation}
\begin{array}{lll}
Cov(X,Y) & = & E[(X-\mu_X)(Y-\mu_y)]
\end{array}
\end{equation}$

La correlación es una medida normalizada:

$\begin{equation}
\begin{array}{lll}
Corr(X,Y) & = & \frac{Cov(X,Y)}{Var(X) Var(Y)}
\end{array}
\end{equation}$



#### Propiedades
$\begin{equation}
\begin{array}{lll} 
Cov(X,Y) & = & Cov(Y,X) \\
Cov(X,X) & = & Var(X)\\
Cov(X+Z,Y) & = & Cov(X,Y) + Cov(Z,Y)\\
Cov(\sum_i \limits X_i,Y) & = & \sum_i \limits Cov(X_i,Y)\\
Var(X+Y) & = & Var(X) + Var(Y) + 2Cov(X,Y)\\
Var(\sum_i \limits X_i) & = & 
\sum_i \limits Var(X_i) + \sum_i \limits \sum_{j\neq i} \limits Cov(X_i,X_j)
\end{array}
\end{equation}$

### Ley débil de los grandes números
Sea $\{X_i\}_{i=1}^N$ un conjunto de $N$ v.a. independientes idénticamente distribuidas (iid) de media $\mu$ y varianza finita $\sigma^2$. Entonces, para cualquier $\epsilon >0$ se tiene:
$$\lim_{N \to \infty} P\left\{\middle | \frac{\sum_i X_i}{N} - \mu \,\middle | > \epsilon \right\} = 0$$

### Regla de Bayes
En su formulación con variables aleatorias, la regla de Bayes queda:

$\begin{equation}
\begin{array}{lll}
P(y \mid x)& = & \frac{P(x \mid y) P_Y(y)}{P_X(x)} \,= \, \frac{P(x \mid y) P_Y(y)}{\sum_y P(x\mid y) P_y(y))}\\
\end{array}\end{equation}$



### Variables aleatorias especiales
Existen algunos tipos de variables aleatorias especiales, definidas por que tienen  distribuciones de probabilidad particulares.

#### Distribución Bernoulli
Suponga que se realiza un ensayo cuya salida es "éxito" o "falla". La v.a. entonces tiene sólo dos valores posibles: 1 si "éxito" 0 si no. Si $p$ es la probabilidad de éxito, entonces:

$\begin{equation}
\begin{array}{lll} 
P(X=1) & = &p \qquad y \qquad P(X=0) = 1-p \\
P(X = i) & = & p^i (1-p)^i \,, i=0,1\\
E[X] & = & p\\
Var(X) & = & p(1-p)\\
Asimetria(X) & = & \frac{1-2p}{\sqrt{p(1-p)}}\\
Curtosis(X) & = & \frac{1-3p+3p^2}{p(1-p)}\\
\end{array}
\end{equation}$


#### Distribución Binomial
Suponga que se realizan $N$  ensayos independientes idénticos  Bernoulli  de parámetro $p$. La v.a. X que representa el número $k$ de ensayos existosos entre los $N$ ensayos realizados, se denomina Binomial y cumple:

$\begin{equation}
\begin{array}{lll} 
P(X=k) & = & {N \choose k} p^k (1-p)^{N-k}, \qquad k=0,1,...N \\
E[X] & = & Np\\
Var(X) & = & Np(1-p)\\
Asimetria(X) & = & \frac{1-2p}{\sqrt{np(1-p)}}\\
Curtosis(X) & = & \frac{1-6p(1-p)}{np(1-p)}\\
\end{array}
\end{equation}$

#### Distribución Exponencial
Sea $X$ v.a. continua, es exponencial de tasa $\lambda$ si su densidad de probabilidad cumple:

$\begin{equation}
\begin{array}{lll} 
f_X(x) & = & \lambda exp(-\lambda x)\qquad, 0 \leq x <\infty\\
F_X(x) & = & 1- exp(-\lambda x)\\
E[X] &= &\frac{1}{\lambda}\\
Var(X)& =& \frac{1}{\lambda^2}\\
Asimetria(X) & = & 2\\
Curtosis(X) & = & 9 \\
\end{array}
\end{equation}$

#### Distribución Normal o Gaussiana
Sea $X$ v.a. continua, es normal de media $\mu$ y varianza $\sigma^2$ y se denota $\cal{N}(\mu,\sigma^2)$ si su densidad de probabilidad cumple:


$\begin{equation}
\begin{array}{lll} 
f_X(x) & = & \frac{1}{\sqrt{2\pi}\sigma} exp(\frac{-(x-\mu)^2}{2\sigma^2})\qquad, -\infty < x <\infty\\
E[X] &= &\mu\\
Var(X)& =& \sigma^2\\
Asimetria(X) & = & 0\\
Curtosis(X) & = & 3 \\
\end{array}
\end{equation}$

Esta es la distribución de probabilidad mas utilizada. Muchas variables se distribuyen aproximadamente normales (altura, peso, satisfacción en el trabajo, etc.).

#### Distribución Normal Estandarizada
$Z \sim \cal{N}(0,1)$ si su densidad de probabilidad cumple:


$\begin{equation}
\begin{array}{lll} 
f_X(x) & = & \frac{1}{\sqrt{2\pi}} exp(\frac{-x^2}{2})\qquad, -\infty < x <\infty\\
E[X] &= &0\\
Var(X)& =&1\\
\end{array}
\end{equation}$

Si $X \sim \cal{N}(\mu,\sigma^2)$, entonces $Z = \frac{X-\mu}{\sigma} \sim \cal{N}(0,1)$

Esta transformación se denomina "Z-score" y se utiliza incluso si la v.a. no cumple el supuesto de normalidad. Se suele denominar estandarización o normalización.



#### Regla Empírica 68-95-99.7 
Si $X \sim \cal{N}(\mu,\sigma^2)$, entonces:

$\begin{equation}
\begin{array}{lll} 
P(\mu-\sigma < X < \mu+\sigma) & = & 0.68\\
P(\mu-2\sigma < X < \mu+2\sigma) & = & 0.95\\
P(\mu-3\sigma < X < \mu+3\sigma) & = & 0.997\\
\end{array}
\end{equation}$



Muy útil para la detección de outliers, bajo el supuesto de normalidad.

#### Teorema del Límite Central
Sean $X_1,...X_N$ v.a. iid de media $\mu$ y varianza $\sigma^2$ entonces
$S_N=X_1+...X_N$ cumple:

$$ \lim_{N \to \infty}P\left( \frac{S_N - N\mu}{\sigma\sqrt N} \leq z\right) = F_Z(z) 
\qquad Z \sim \cal{N}(0,1)$$


En particular $m= \frac{S_N}{N}= \frac{\sum_{i=1}^N \limits X_i}{N}$ cumple:

$$ \lim_{N \to \infty}P\left( \frac{m - \mu}{\sigma/\sqrt N} \leq z\right) = F_Z(z)  $$

Es decir que es posible aproximar la distribución de $m$ por ${\cal{N}}(\mu,\frac{\sigma^2}{N})$
