### Definicin Formal de Probabilidad
Una Probabilidad $P$ es una funcin que cumple:
\begin{equation}
\begin{array}{cc} P:& \Omega \to [0,1]\\
&A \to P(A)
\end{array}
\end{equation}
que cumple lo siguiente:

$\begin{equation}
\begin{array}{ll} 1. & \forall A \in \Omega, P(A) \geq 0\\
2. &  P(\Omega) = 1\\
3. & Sean \,\, A_1,...,A_n \in \Omega: A_i \cap A_j =\phi, \forall i,j=1...n => P(A_1\cup...\cup A_n) = P(A_1)+...+P(A_n)\\
\end{array}
\end{equation}$

### Propiedades
$\begin{equation}
\begin{array}{ll} 
1. & P(A) = 1 - P(A^C)\\
2. &  P(\phi) = 0\\
3. & Si A \subset B => P(A) \leq P(B)\\
4. & P(A) \leq 1\\
5. & \forall A,B \in \Omega\,\, P(A \cup B) = P(A) + P(B) - P(A \cap B)
\end{array}
\end{equation}$

### Probabilidades condicionales
Sean $A, B \in \Omega$, si $P(B)>0$, se define:

$\begin{equation}
\begin{array}{lll} 
P(A\mid B) & = &\frac{P(A \cap B)}{P(B)}
\end{array}\end{equation}$

Si adems se cumple que $P(A)>0$, se tiene:

$\begin{equation}
\begin{array}{lll}
P(A\cap B)& = &P(A\mid B)P(B) = P(B\mid A)P(A)
\end{array}\end{equation}$

Esto se puede extender a mas eventos:

$\begin{equation}
\begin{array}{lll}
P(A\cap B \cap C)& = &P(A)P(B\mid A) P(C\mid A \cap B) 
\end{array}\end{equation}$
siempre que $P(A), P(A\cap B)> 0$

### Independencia
Cmo determinar que dos eventos son independientes?

Sean $A, B \in \Omega$ con $P(A),P(B) >0$, si cualquiera de las siguientes condiciones se cumple:

$\begin{equation}
\begin{array}{llll}
1. &P(A\cap B)& = &P(A)P(B)\\ 
2. &P(A\mid B) & = & P(A)\\
3. &P(B\mid A) & = & P(B)\\
\end{array}\end{equation}$

se dice que $A$ es independiente de $B$.

### Independencia Condicional
Cmo determinar que dos eventos $A$ y $B$ son independientes dados un tercer evento $C$?

Sean $A, B, C \in \Omega$ con $P(A\cap C),P(B\cap C), P(C)>0$, si cualquiera de las siguientes condiciones se cumple:

$\begin{equation}
\begin{array}{llll}
1. &P(A\cap B \mid C)& = &P(A\mid C)P(B\mid C)\\ 
2. &P(A\mid B \cap C) & = & P(A \mid C)\\
3. &P(B\mid A \cap C) & = & P(B \mid C)\\
\end{array}\end{equation}$

se dice que $A$ es independiente de $B$ condicionalmente a $C$.

### Ley de las Probabilidades totales
Consideremos ahora una particin del espacio muestral $\{B_i\}_{i=1,..n}$, es decir:

$\begin{equation}
\begin{array}{lll}
\bigcup_{i=1}^n \limits  B_i = \Omega, & & B_i\cap B_j = \phi \qquad \forall i,j=1,..n, i\neq j
\end{array}\end{equation}$

entonces:

$\begin{equation}
\begin{array}{lll}
P(A) & = & \sum_{i=1}^n \limits P(A\mid B_i) P(B_i)
\end{array}\end{equation}$

Ejemplo: Sea $A$ = \{Pedro acepta trabajo en Chile\}
y $B_i$ = \{Oferta de trabajo para Pedro en region $i$ \}, $i=1,...,15$ 






### Regla de Bayes
A partir de la deficin de probabilidades condicionales se puede deducir la regla de Bayes:

Sean $A, B \in \Omega$ tales que $P(A),P(B)>0$, entonces 

$\begin{equation}
\begin{array}{lll}
P(B\mid A)& = & \frac{P(A\mid B) P(B)}{P(A)}
\end{array}\end{equation}$

$P(B)$ se denomina probabilidad a priori, $P(B \mid A)$ probabilidad a posteriori

### Teorema de Bayes
Consideremos una particin del espacio muestral $\{B_i\}_{i=1,..n}$, entonces:

$\begin{equation}
\begin{array}{lll}
P(B_i \mid A) & = & \frac{P(A\mid B_i) P(B_i)}{P(A)}
& = & \frac{P(A\mid B_i) P(B_i)}{\sum_{i=1}^n \limits P(A\mid B_i) P(B_i)}
\end{array}\end{equation}$

### Variables Aleatorias
Una variable aleatoria es una funcin que asigna un nmero a cada elemento en el espacio muestral de un experimento aleatorio. 
Se denomina Variable Aleatoria Discreta si los valores son enteros:

$\begin{equation}
\begin{array}{cc} X:& \Omega \to \cal{N}\\
&\omega \to X(\omega)
\end{array}
\end{equation}$

y Variable Aleatoria Continua si los valores son reales:

$\begin{equation}
\begin{array}{cc} X:& \Omega \to \cal{R}\\
&\omega \to X(\omega)
\end{array}
\end{equation}$


Ejemplos: 

$\begin{equation}
\begin{array}{ll}

\end{array} \right .\\
&\\

&\omega \to X(\omega)= t \\
\end{array}
\end{equation}$

donde $t$ es el tiempo de falla de una mquina y $\omega$ es el evento de falla con todas sus caractersticas no necesariamente observables.

### Funcin de distribucin de probabilidad acumulada
Es la funcin $F_X$ que devuelve la probabilidad de que una v.a. sea menor o igual que un valor.

En el caso de las v.a. discretas se calcula como:

$\begin{equation}
\begin{array}{ll} 
P(X \leq x) = F_X(x) & = \sum_{t=1}^x \limits P(X = t)
\end{array}
\end{equation}$


En el caso de las v.a. continuas se calcula como:

$\begin{equation}
\begin{array}{llll} 
P(X \leq x)  = &F_X(x)& = &\int_{-\infty}^x f(x)dx\\
&f_X(x)& = & \frac{d}{dx} F_X(x)
\end{array}
\end{equation}$

Y una manera alternativa de obtener la probabilidad:

$\begin{equation}
\begin{array}{l}
P(a \leq X \leq b) = F_X(b) - F_X(a)\\
\end{array}
\end{equation}$

### Esperanza
La esperanza o  valor esperado de una v.a. $X$ se denota $E[X]$ y se calcula como:
$\begin{equation}
\begin{array}{ll} 
E[X] = 
\left\{\begin{array}{ll} \sum_i x_i P(X=x_i) & si\,X\, discreta\\
                          \int x f_X(x)dx & si\,X\, continua\\
\end{array} \right .\\
\end{array}
\end{equation}$


Consideremos $g$ una funcin a valores reales, entonces:

$\begin{equation}
\begin{array}{lll}
E[g(X)] & = & 
\left\{\begin{array}{ll} \sum_i g(x_i) P(X=x_i) & si\,X\, discreta\\
                          \int g(x) f_X(x)dx & si\,X\, continua\\
\end{array}\right .\\
\end{array}
\end{equation}$

Para el caso especial de $g(x) = x^n$ se define el n-simo momento de X como:

$\begin{equation}
\begin{array}{lll} 
E[X^n] & = & 
\left\{\begin{array}{ll} \sum_i x_i^n P(X=x_i) & si\,X\, discreta\\
                          \int x^n f_X(x)dx & si\,X\, continua\\
\end{array}\right .\\
\end{array}
\end{equation}$

La esperanza es el primer momento y se denota $\mu$.

#### Propiedades
Sean $a,b \in \cal{R}$ entonces:

$\begin{equation}
\begin{array}{lll} 
E[aX+b] & = & aE[X] + b \\
E[X + Y] & = & E[X] + E[Y]\\
\end{array}
\end{equation}$

### Varianza y covarianza
La varianza mide la variacin de la v.a. entorno a la esperanza o media $\mu$, y se define como

$\begin{equation}
\begin{array}{ll} 
Var(X) = E[(X-\mu)^2] = E[X^2] - \mu^2
\end{array}
\end{equation}$

Se cumple que:

$\begin{equation}
\begin{array}{ll} 
Var(aX+b) = a^2 Var(X)
\end{array}
\end{equation}$

Se define adems la desviacin estndar $\sigma = \sqrt{Var(X)}$ 


La covarianza mide la relacin (lineal) que hay entre dos v.a. $X$ e $Y$. Si denotamos $\mu_X = E[X]$ y $ \mu_Y= E[Y]$ entonces:

$\begin{equation}
\begin{array}{lll}
Cov(X,Y) & = & E[(X-\mu_X)(Y-\mu_y)]
\end{array}
\end{equation}$

La correlacin es una medida normalizada:

$\begin{equation}
\begin{array}{lll}
Corr(X,Y) & = & \frac{Cov(X,Y)}{Var(X) Var(Y)}
\end{array}
\end{equation}$



#### Propiedades
$\begin{equation}
\begin{array}{lll} 
Cov(X,Y) & = & Cov(Y,X) \\
Cov(X,X) & = & Var(X)\\
Cov(X+Z,Y) & = & Cov(X,Y) + Cov(Z,Y)\\
Cov(\sum_i \limits X_i,Y) & = & \sum_i \limits Cov(X_i,Y)\\
Var(X+Y) & = & Var(X) + Var(Y) + 2Cov(X,Y)\\
Var(\sum_i \limits X_i) & = & 
\sum_i \limits Var(X_i) + \sum_i \limits \sum_{j\neq i} \limits Cov(X_i,X_j)
\end{array}
\end{equation}$

### Ley dbil de los grandes nmeros
Sea $\{X_i\}_{i=1}^N$ un conjunto de $N$ v.a. independientes idnticamente distribuidas (iid) de media $\mu$ y varianza finita $\sigma^2$. Entonces, para cualquier $\epsilon >0$ se tiene:
$$\lim_{N \to \infty} P\left\{\middle | \frac{\sum_i X_i}{N} - \mu \,\middle | > \epsilon \right\} = 0$$

En su formulacin con variables aleatorias, la regla de Bayes queda:

\begin{array}{lll}
P(y \mid x)& = & \frac{P(x \mid y) P_Y(y)}{P_X(x)} \,= \, \frac{P(x \mid y) P_Y(y)}{\sum_y P(x\mid y) P_y(y))}\\
\end{array}\end{equation}$



### Variables aleatorias especiales
Existen algunos tipos de variables aleatorias especiales, definidas por que tienen  distribuciones de probabilidad particulares.

#### Distribucin Bernoulli
Suponga que se realiza un ensayo cuya salida es "xito" o "falla". La v.a. entonces tiene slo dos valores posibles: 1 si "xito" 0 si no. Si $p$ es la probabilidad de xito, entonces:

$\begin{equation}
\begin{array}{lll} 
P(X=1) & = &p \qquad y \qquad P(X=0) = 1-p \\
P(X = i) & = & p^i (1-p)^i \,, i=0,1\\
E[X] & = & p\\
Var(X) & = & p(1-p)\\
\end{array}
\end{equation}$


#### Distribucin Binomial
Suponga que se realizan $N$  ensayos independientes idnticos  Bernoulli  de parmetro $p$. La v.a. X que representa el nmero $k$ de ensayos existosos entre los $N$ ensayos realizados, se denomina Binomial y cumple:

$\begin{equation}
\begin{array}{lll} 
P(X=k) & = & {N \choose k} p^k (1-p)^{N-k}, \qquad k=0,1,...N \\
E[X] & = & Np\\
Var(X) & = & Np(1-p)\\
\end{array}
\end{equation}$

#### Distribucin Normal o Gaussiana
Sea $X$ v.a. continua, es normal de media $\mu$ y varianza $\sigma^2$ y se denota $\cal{N}(\mu,\sigma^2)$ si su densidad de probabilidad cumple:


$\begin{equation}
\begin{array}{lll} 
f_X(x) & = & \frac{1}{\sqrt{2\pi}\sigma} exp(\frac{-(x-\mu)^2}{2\sigma^2})\qquad, -\infty < x <\infty\\
E[X] &= &\mu\\
Var(X)& =& \sigma^2\\
\end{array}
\end{equation}$

#### Distribucin Normal Estandarizada
$Z \sim \cal{N}(0,1)$ si su densidad de probabilidad cumple:


$\begin{equation}
\begin{array}{lll} 
f_X(x) & = & \frac{1}{\sqrt{2\pi}} exp(\frac{-x^2}{2})\qquad, -\infty < x <\infty\\
E[X] &= &0\\
Var(X)& =&1\\
\end{array}
\end{equation}$

Si $X \sim \cal{N}(\mu,\sigma^2)$, entonces $Z = \frac{X-\mu}{\sigma} \sim \cal{N}(0,1)$

Esta transformacin se denomina "Z-score" y se utiliza incluso si la v.a. no cumple el supuesto de normalidad. Se suele denominar estandarizacin o normalizacin.

#### Teorema del Lmite Central
Sean $X_1,...X_N$ v.a. iid de media $\mu$ y varianza $\sigma^2$ entonces
$S_N=X_1+...X_N$ cumple:

$$ \lim_{N \to \infty}P\left( \frac{S_N - N\mu}{\sigma\sqrt N} \leq z\right) = F_Z(z) 
\qquad Z \sim \cal{N}(0,1)$$


En particular $m= \frac{S_N}{N}= \frac{\sum_{i=1}^N \limits X_i}{N}$ cumple:

$$ \lim_{N \to \infty}P\left( \frac{m - \mu}{\sigma/\sqrt N} \leq z\right) = F_Z(z)  $$

Es decir que es posible aproximar la distribucin de $m$ por ${\cal{N}}(\mu,\frac{\sigma^2}{N})$
